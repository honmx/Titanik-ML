# Feature Importances

```python
def age_group(age):
    if pd.isnull(age):
        return ''
    elif age < 10:
        return 'child'
    elif 10 <= age < 20:
        return 'teenager'
    elif 20 <= age < 60:
        return 'adult'
    else:
        return 'old'


def fare_group(fare):
    if fare < 10:
        return 'cheap'
    elif 10 <= fare <= 50:
        return 'normal'
    elif fare > 50:
        return 'expensive'
    else:
        return ''

# Добавляем новый столбец AgeGroup, в котором содержится возрастная категория, которой принадлежит пассажир
df['AgeGroup'] = df['Age'].apply(age_group)
# Добавляем новый столбец FareGroup, в котором содержится категория билета, который был у пассажира
df['FareGroup'] = df['Fare'].apply(fare_group)
# Добавляем новый столбец Relatives, в котором содержится общее кол-во родственников
df['Relatives'] = df['SibSp'] + df['Parch']
```

# Модель

```python
X = pd.get_dummies(df.drop(columns=['Survived']), drop_first=True)
y = df['Survived']
```

```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
```

### Модель k-neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

alg_ngbh = KNeighborsClassifier(n_neighbors=3)
scores = cross_val_score(alg_ngbh, X, y, cv=cv, n_jobs=-1)
print("Accuracy (k-neighbors): {}/{}".format(scores.mean(), scores.std()))
```

### Модель SVC

```python
from sklearn.svm import SVC

alg_svm = SVC(C=1.0)
scores = cross_val_score(alg_svm, X, y, cv=cv, n_jobs=-1)
print("Accuracy (svm): {}/{}".format(scores.mean(), scores.std()))
```

### Модель logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

def custom_accuracy_scorer(estimator, X_test, y_test):
    predictions = estimator.predict(X_test)

    # Преобразование вероятностей в классы
    predictions[predictions > 0.5] = 1
    predictions[predictions <= 0.5] = 0

    return accuracy_score(y_test, predictions)

# Инициализация модели логистической регрессии
logistic_model = LogisticRegression(random_state=1)

# Оценка модели с использованием кросс-валидации
cv_scores = cross_val_score(logistic_model, X, y, cv=5, n_jobs=-1, scoring=custom_accuracy_scorer)

# Вывод результатов
print("Accuracy (Logistic Regression): {}/{}".format(cv_scores.mean(), cv_scores.std()))
```

### Модель RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

random_forest_model = RandomForestClassifier(random_state=1)

random_forest_params = [{
    "n_estimators": [350, 400, 450],
    "min_samples_split": [6, 8, 10],
    "min_samples_leaf": [1, 2, 4]
}]

grid_search = GridSearchCV(random_forest_model, random_forest_params, cv=5, refit=True, verbose=1, n_jobs=-1)

grid_search.fit(X, y)

best_random_forest_model = grid_search.best_estimator_
print("Best Accuracy: {:.2f} with params {}".format(grid_search.best_score_, grid_search.best_params_))
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, classification_report

y_pred = cross_val_predict(random_forest_model, X, y, cv=cv)


conf_matrix = confusion_matrix(y, y_pred)
tn, fp, fn, tp = conf_matrix.ravel()


print(conf_matrix)
print(f'Предсказала, что погибнет, а по факту не погиб = {fp}\n'
      f'Предсказала, что не погибнет, а по факту погиб = {fn}\n'
      f'Предсказала, что не погибнет, и по факту не погиб = {tn}\n'
      f'Предсказала, что погибнет, и по факту погиб = {tp}\n')


class_report = classification_report(y, y_pred)
print(f'Classification Report:\n{class_report}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(random_forest_model, X, y, cv=cv)

plt.figure(figsize=(12, 5))
sns.boxplot(x=np.arange(1, cv.get_n_splits() + 1), y=accuracies)
plt.xlabel('Фолд')
plt.ylabel('Точность')
plt.xticks(np.arange(0, cv.get_n_splits(), 1), [f'Фолд {i+1}' for i in range(cv.get_n_splits())])
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    random_forest_model, X, y, cv=cv, n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10)
)

mean_train_scores = np.mean(train_scores, axis=1)
mean_test_scores = np.mean(test_scores, axis=1)
std_train_scores = np.std(train_scores, axis=1)
std_test_scores = np.std(test_scores, axis=1)

plt.figure(figsize=(12, 6))
plt.xlabel('Размер обучающей выборки')
plt.ylabel('Точность')
plt.ylim(0.0, 1.1)
plt.grid()

plt.fill_between(train_sizes, mean_train_scores - std_train_scores, mean_train_scores + std_train_scores,
                 alpha=0.1, color="r", label="Обучающая выборка (±std)")

plt.fill_between(train_sizes, mean_test_scores - std_test_scores, mean_test_scores + std_test_scores,
                 alpha=0.1, color="g", label="Тестовая выборка (±std)")

plt.plot(train_sizes, mean_train_scores, 'o-', color="r", label="Обучающая выборка")
plt.plot(train_sizes, mean_test_scores, 'o-', color="g", label="Тестовая выборка")
plt.legend(loc="best")
plt.show()
```

Вывод: Был проведен анализ датасета, на основе которого были созданы новые признаки. Были созданы 4 модели, лучшая из которых - RandomForest с accuracy = 0.84. Можно сказать, что модель достаточно точно определяет исход, т.к. мы имеем высокие показатели tn и tp и низкие показатели fn и fp.

```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = Sequential()
model.add(Dense(40, activation="relu", input_shape=(X_train.shape[1],)))
model.add(Dropout(0.2))
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

train_history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.3)
loss, accuracy = model.evaluate(X_test, y_test)

print(f'test loss {loss:.3f}, test accuracy {accuracy:.3f}')
plt.plot(train_history.history['accuracy'], label='accuracy на трейне') 
plt.plot(train_history.history['val_accuracy'], label='accuracy на валидации')
plt.legend(loc=(1.05, 0.5))
```
